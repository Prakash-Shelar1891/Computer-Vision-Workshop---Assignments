```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
rectangle = np.zeros((300, 300), dtype = 'uint8')
rectangle = cv2.cvtColor(rectangle, cv2.COLOR_BGR2RGB)
cv2.rectangle(rectangle, (25, 25), (275, 275), (255, 255, 255), -1)
plt.imshow(rectangle)
plt.show()
```

```python
circle = np.zeros((300, 300), dtype='uint8')
circle = cv2.cvtColor(circle, cv2.COLOR_BGR2RGB)
cv2.circle(circle, (150, 150), 150, (255, 255, 255), -1)
plt.imshow(circle)
plt.show()
```

```python
#Bitwise XOR
bitXOR = cv2.bitwise_xor(rectangle, circle)
plt.imshow(bitXOR)
plt.show()
```

```python
#Bitwise NOT
bitNOT = cv2.bitwise_not(rectangle, circle)
plt.imshow(bitNOT)
plt.show()
```
