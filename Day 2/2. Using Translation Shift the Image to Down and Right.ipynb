```python
import cv2
import matplotlib.pyplot as plt
import imutils
```

```python
img = cv2.imread('sample2.jpg')
```

```python
%matplotlib inline
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
#Shifting Down
down = imutils.translate(img,0,100)
plt.imshow(down)
plt.show()
```

```python
#Shifting Right
right = imutils.translate(img,50,0)
plt.imshow(right)
plt.show()
```
